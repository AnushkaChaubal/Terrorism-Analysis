```python
df=data.groupby(['region_txt'])['eventid'].size().reset_index(name='count')
```

```python
df_year = data.groupby(['iyear','region_txt']).size().reset_index(name='Attack Count')

# Step 3: Create the graph
plt.figure(figsize=(10, 6))

lines=[]
labels=[]

# Plotting lines for each region
for region in df['region_txt']:
    region_data = df_year[df_year['region_txt'] == region]
    plt.plot(region_data['iyear'], region_data['Attack Count'],label=region)


# Graph labels and title
plt.xlabel('Year')
plt.ylabel('Number of Terror Attacks')
plt.title('Terror Attacks Over Time by Region')
plt.legend()

# Display the graph
plt.show()
```

```python
print(df)
```

```python
avg=df['count'].mean()
print(avg)
```

```python
req_regions=[]
for index, row in df.iterrows():
    if row['count'] > avg:
        req_regions.append(row['region_txt'])
print(req_regions)
```

```python
plt.figure(figsize=(15,8))
plt.bar(df['region_txt'], df['count'], label='All Categories')
plt.bar(req_regions, df.loc[df['region_txt'].isin(req_regions), 'count'],
        label='Selected Categories (Count > global average)')

plt.axhline(y=avg, color='r', linestyle='--', label='Mean Count')

plt.xlabel('Region')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.title('Count by Region')
plt.legend()

plt.show()
```

```python
pivot_table = pd.pivot_table(df_year, values='Attack Count', index='iyear', columns='region_txt', aggfunc='sum')
print(pivot_table)
```

```python
globalMean=pivot_table.mean(axis=1)
pivot_table['global_avg']=globalMean
```

```python
terror_regions=['Middle East & North Africa','South America','South Asia','Sub-Saharan Africa','Western Europe']
df_terror= data[data['region_txt'].isin(terror_regions)]
df_terror.head()
```

```python
df_byYear = df_terror.groupby(['iyear','region_txt']).size().reset_index(name='Attack Count')

# Step 3: Create the graph
plt.figure(figsize=(10, 6))

# Plotting lines for each region
for region in terror_regions:
    region_data = df_byYear[df_byYear['region_txt'] == region]
    plt.plot(region_data['iyear'], region_data['Attack Count'], label=region)

# Graph labels and title
plt.xlabel('Year')
plt.ylabel('Number of Terror Attacks')
plt.title('Terror Attacks Over Time by most attacked Regions')
plt.legend()

# Display the graph
plt.show()
```
